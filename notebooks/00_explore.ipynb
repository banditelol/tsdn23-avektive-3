## Match All Shodan to NVD

```python
ip_cve = pd.read_csv("../data/ip_cve_shodan.csv")
shodan_ips = list(ip_cve["ip"].unique())
shodan_cves = list(ip_cve["cve_id"].unique())
```

```python
cve_col_map = {
    "id": "id",
    "publishedDate": "published_date",
    "cve.problemtype.problemtype_data": "problem_type",
    "cve.description.description_data": "description",
    "impact.baseMetricV2.cvssV2.vectorString": "cvssv2_vectortring",
    "impact.baseMetricV2.cvssV2.accessVector": "cvssv2_accessvector",
    "impact.baseMetricV2.cvssV2.baseScore": "cvssv2_base_score",
    "impact.baseMetricV3.cvssV3.vectorString": "cvssv3_vectorstring",
    "impact.baseMetricV3.cvssV3.attackVector": "cvssv3_attackvector",
    "impact.baseMetricV3.cvssV3.baseScore": "cvssv3_base_score",
}


shodan_cve_df = (
    pd.read_parquet(
        "../data/nvdcve-1.1-all.parquet",
        filters=[("id", "in", shodan_cves)],
        columns=list(cve_col_map.keys()),
    )
    .rename(columns=cve_col_map)
    .assign(
        problem_type=lambda df: df["problem_type"]
        .str[0]
        .str["description"]
        .str[0]
        .str["value"],
        description=lambda df: df["description"].str[0].str["value"],
    ).set_index("id")
)
shodan_cve_df.to_csv("../data/shodan_cve.csv")
shodan_cve_df
```

```python
epss_df = pd.read_csv("../data/epss.csv").set_index("cve")
shodan_epss_df = shodan_cve_df.merge(epss_df, how="left", left_index=True, right_index=True)
shodan_epss_df.to_csv("../data/shodan_cve_epss.csv")
```

```python
kev_df = pd.read_csv("../data/cisa-kev.csv")[["cveID","vendorProject","product","vulnerabilityName","knownRansomwareCampaignUse"]].set_index("cveID")
shodan_kev_df = shodan_epss_df.merge(kev_df, how="left",left_index=True,right_index=True)
shodan_kev_df.to_csv("../data/shodan_cve_epss_kev.csv")
```
